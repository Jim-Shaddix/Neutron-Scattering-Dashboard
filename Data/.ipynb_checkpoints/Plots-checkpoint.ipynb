### Imports

```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.plotly as py
import matplotlib.pyplot as plt
np.set_printoptions(precision=2, linewidth=150)
```

### Read In Data

z-data
* 164 rows
* 58 elements per row

* finding information on the file

```python
flag_totals = False # print summary statistics for groups of numbers
flag_elements = False # print groups of nnumbers
with open("NCCF_elastic_folded.xyie") as fd: 
    count = 0 
    for line in fd: 
        if len(line) > 30: 
            line_elements = line.strip().split()
            if flag_totals:
                print("Element Count:", len(line_elements)) 
            if flag_elements:
                elements = np.array(line_elements, dtype=np.float128)
                elements = np.hstack([elements, [0,0]]).reshape((6,10))
                print(elements,end="\n\n")
            count += 1; 
    if flag_totals:
        print("total:",count) 
```

```python
df = pd.read_csv("data_from_NaCaCo2F7_for_Plotly.csv",names=["x","y","z"])
df.shape
```

### Data Cross Sections

* y-data

```python
%matplotlib inline
plt.scatter(range(len(df.y)), df.y, s=1, marker='o');
plt.title("Y-Data");
```

* x vs. y data

```python
%matplotlib inline
plt.scatter(df.y,
            df.x, 
            s=5,
            c=np.arange(len(df.x)) // len(df.x.unique()), 
            marker='o');
plt.title("X-Data");
```

* z-data

```python
plt.scatter(range(len(df.z)), df.z,s=1)
plt.title("Z-Data");
```

# Plots

#### Scatter

```python
data = [go.Scatter3d(x=df.x, y=df.y, z=df.z,
                     mode="markers",
                     marker = dict(
                         color = '#FFBAD2',
                         line = dict(width = 0.01)
                     )
                    )]
```

```python
py.iplot(go.Figure(data))
```

### Heat Map

```python
trace = go.Heatmap(
    z=np.array(df.z).reshape(-1, (len(df.x.unique()))),
    colorscale='Viridis'
)
data=[trace]
py.iplot(data)
```

### Surface Plot

```python
x_unique = df.x.unique()
y_unique = df.y.unique()
len(x_unique), len(y_unique)
```

```python
X, Y = np.meshgrid(x_unique,y_unique)
X.shape, Y.shape
```

```python
Z = np.array(df.z).reshape((len(y_unique), len(x_unique)))
```

```python
data = [go.Surface(x=X, y=Y, z=Z)]
```

```python
py.iplot(go.Figure(data=data))
```
